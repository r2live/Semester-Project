```python
#Dropping the "Type 1" and "Type 2" columns, and splitting off the "Tier" column, which contains the labels

df = df.drop('Type 1', axis = 1)
df = df.drop('Type 2', axis = 1)

y = df.pop('Tier')

df.insert(0, '#_of_x4_Weaknesses', 0)
df.insert(0, '#_of_x2_Weaknesses', 0)
df.insert(0, '#_of_Resistances', 0)
df.insert(0, '#_of_Weaknesses', 0)


#Loop to populate the "#_of_Weaknesses", "#_of_Resistances", "#_of_x2_Weaknesses", and "#_of_x4_Weaknesses" columns
for index, row in df.iterrows():
    weak = 0
    res = 0
    x2 = 0
    x4 = 0
    for i in range(4, 22):
        if row[i] == 2.0:
            weak += 1
            x2 += 1
        elif row[i] == 4.0:
            weak += 1
            x4 += 1
        elif row[i] == 0 or row[i] == 0.25 or row[i] == 0.5:
            res += 1
        else:
            pass
    df.at[index, '#_of_Weaknesses'] = weak
    df.at[index, '#_of_Resistances'] = res
    df.at[index, '#_of_x2_Weaknesses'] = x2
    df.at[index, '#_of_x4_Weaknesses'] = x4
    
#print(df.columns.values)
    
    
```

```python
#Splitting the data into training and testing data and trying out various regressors/ensemble methods, and printing out
#mean squared error and accuracy scores for each one

X_train, X_test, y_train, y_test = train_test_split(df, y, test_size = 0.2, random_state = 0)

from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import accuracy_score

from sklearn.linear_model import LinearRegression
lr = LinearRegression(normalize = True)
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)

print("Linear Regression:")
print(mean_squared_error(y_test, y_pred.round(decimals = 0)))
print(accuracy_score(y_test, y_pred.round(decimals = 0)))
print("----------------")

from sklearn.linear_model import LogisticRegression
logr = LogisticRegression()
logr.fit(X_train, y_train)
y_pred = logr.predict(X_test)

print("Logistic Regression:")
print(mean_squared_error(y_test, y_pred.round(decimals = 0)))
print(accuracy_score(y_test, y_pred.round(decimals = 0)))
print("----------------")

from sklearn.tree import DecisionTreeRegressor
regr = DecisionTreeRegressor(max_depth = 3)
regr.fit(X_train, y_train)
y_pred = regr.predict(X_test)

print("Decision Tree Regression:")
print(mean_squared_error(y_test, y_pred.round(decimals = 0)))
print(accuracy_score(y_test, y_pred.round(decimals = 0)))
print("----------------")

from sklearn.neighbors import KNeighborsRegressor
knnreg = KNeighborsRegressor(n_neighbors = 28, p = 1)
knnreg.fit(X_train, y_train)
y_pred = knnreg.predict(X_test)

print("KNN Regression:")
print(mean_squared_error(y_test, y_pred.round(decimals = 0)))
print(accuracy_score(y_test, y_pred.round(decimals = 0)))
print("----------------")

from sklearn.svm import SVR
svreg = SVR()
svreg.fit(X_train, y_train)
y_pred = svreg.predict(X_test)

print("Support Vector Regression:")
print(mean_squared_error(y_test, y_pred.round(decimals = 0)))
print(accuracy_score(y_test, y_pred.round(decimals = 0)))
print("----------------")

from sklearn.linear_model import SGDRegressor
sgd = SGDRegressor(max_iter = 500)
sgd.fit(X_train, y_train)
y_pred = sgd.predict(X_test)

print("Stochastic Gradient Descent Regression:")
print(mean_squared_error(y_test, y_pred.round(decimals = 0)))
print(accuracy_score(y_test, y_pred.round(decimals = 0)))
print("----------------")

from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(max_depth = 3)
rfr.fit(X_train, y_train)
y_pred = rfr.predict(X_test)

print("Random Forest Regression:")
print(mean_squared_error(y_test, y_pred.round(decimals = 0)))
print(accuracy_score(y_test, y_pred.round(decimals = 0)))
print("----------------")

from sklearn.linear_model import Lasso
lasso = Lasso(alpha = 0.1)
lasso.fit(X_train, y_train)
y_pred = lasso.predict(X_test)

print("Lasso Regression:")
print(mean_squared_error(y_test, y_pred.round(decimals = 0)))
print(accuracy_score(y_test, y_pred.round(decimals = 0)))
print("----------------")

from sklearn.linear_model import ElasticNet
elastic = ElasticNet(random_state = 0)
elastic.fit(X_train, y_train)
y_pred = elastic.predict(X_test)

print("Elastic Net Regression:")
print(mean_squared_error(y_test, y_pred.round(decimals = 0)))
print(accuracy_score(y_test, y_pred.round(decimals = 0)))
print("----------------")

from sklearn.linear_model import Ridge
ridge = Ridge(alpha = 1.0)
ridge.fit(X_train, y_train)
y_pred = ridge.predict(X_test)

print("Ridge Regression:")
print(mean_squared_error(y_test, y_pred.round(decimals = 0)))
print(accuracy_score(y_test, y_pred.round(decimals = 0)))
print("----------------")

from sklearn.ensemble import AdaBoostRegressor
adareg = AdaBoostRegressor(n_estimators = 50)
adareg.fit(X_train, y_train)
y_pred = adareg.predict(X_test)

print("AdaBoost Regression:")
print(mean_squared_error(y_test, y_pred.round(decimals = 0)))
print(accuracy_score(y_test, y_pred.round(decimals = 0)))
print("----------------")

from sklearn.ensemble import GradientBoostingRegressor
gbreg = GradientBoostingRegressor(n_estimators = 200, learning_rate = 0.8, max_depth = 3, random_state = 0)
gbreg.fit(X_train, y_train)
y_pred = gbreg.predict(X_test)

print("Gradient Boosting Regression:")
print(mean_squared_error(y_test, y_pred.round(decimals = 0)))
print(accuracy_score(y_test, y_pred.round(decimals = 0)))
print("----------------")

#scores = cross_val_score(knnreg, df, y, cv = 5)
#print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))


```
